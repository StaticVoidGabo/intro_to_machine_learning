```python
import datetime
sales = pd.read_csv("data/competitive-data-science-predict-future-sales/sales_train.csv.gz")
sales['date'] = sales['date'].apply(lambda x:datetime.datetime.strptime(x, '%d.%m.%Y'))
```

## Parte 1. Venta total

Predecir la venta total en cada d√≠a de dos ultimas semanas del mes de octubre de 2015.

```python
total_daily_sales = sales.groupby('date', as_index=False)[['item_cnt_day']].sum()\
    .rename({'item_cnt_day': 'total_sales'}, axis=1)
```

```python
total_daily_sales.set_index('date').plot()
```
